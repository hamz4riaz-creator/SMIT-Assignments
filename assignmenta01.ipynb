# **Q7:** `while loop` ka use karke user se tab tak numbers input lein jab tak wo `0` enter na karde. End mein saare numbers ka total dikhayein.

````python
# type your code here
total = 0

print("Numbers enter karein (Rukne ke liye '0' dabayein):")

# Infinite loop shuru karein
while True:
    number = int(input("Number: "))
    
    # Agar user 0 enter kare toh loop tod dein
    if number == 0:
        break
    
    # Number ko total mein jodein
    total += number

print(f"\nSaare numbers ka total sum hai: {total}")
````

# **Q8:** Ek list `numbers = [12, 45, 7, 23, 56, 90, 11]` mein se sabse bada (Maximum) number find karein loop ke zariye.

````python
# type your code here
numbers = [12, 45, 7, 23, 56, 90, 11]

# Maan lete hain ki pehla number hi sabse bada hai
max_num = numbers[0]

# List ke har number ko check karein
for num in numbers:
    if num > max_num:
        max_num = num  # Agar naya number bada mila, toh max_num ko update karein

print(f"List mein sabse bada number hai: {max_num}")
````

# **Q9:** User se ek string input lein (e.g., 'Python') aur loop ka use karke us string ko reverse print karein.

````python
# type your code here
text = input("Ek word likhein: ")

reversed_text = ""

# Loop ko piche se shuru karein
# len(text) - 1 se lekar 0 tak (-1 step ke saath)
for i in range(len(text) - 1, -1, -1):
    reversed_text += text[i]

print(f"Original: {text}")
print(f"Reversed: {reversed_text}")
````

# **Q10:** 1 se 50 tak ke numbers check karein. Agar number 3 se divide hota hai to 'Fizz' print karein, agar 5 se hota hai to 'Buzz' print karein, aur agar dono se hota hai to 'FizzBuzz' print karein.

````python
# type your code here
# 1 se 50 tak loop chalayein
for i in range(1, 51):
    # Sabse pehle check karein agar 3 aur 5 dono se divide hota hai
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    # Phir check karein agar sirf 3 se hota hai
    elif i % 3 == 0:
        print("Fizz")
    # Phir check karein agar sirf 5 se hota hai
    elif i % 5 == 0:
        print("Buzz")
    # Agar kisi se nahi hota, toh number print karein
    else:
        print(i)
````

# **Q11:** Ek list di gayi hai `items = [1, 2, 3, 4, 5]`. Loop ka use karke ek nayi list banayein jis mein har number ka **Square** (e.g., 1, 4, 9...) ho.

````python
# type your code here
items = [1, 2, 3, 4, 5]
squares = []  # Nayi khali list

# Loop ka use karke har number ka square nikaalein
for num in items:
    square_value = num * num  # Ya num ** 2
    squares.append(square_value)

print(f"Original items: {items}")
print(f"Square items:   {squares}")
````

# **Q12:** User se ek number lein aur check karein ke wo **Prime Number** hai ya nahi (using loops).

````python
# type your code here
num = int(input("Ek number enter karein: "))

is_prime = True

if num <= 1:
    is_prime = False  # 1 aur usse chhote numbers prime nahi hote
else:
    # 2 se lekar num/2 tak check karein (kyunki usse bada divisor nahi ho sakta)
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break  # Agar divide ho gaya, toh aage check karne ki zaroorat nahi

# Result print karein
if is_prime:
    print(f"{num} ek Prime Number hai!")
else:
    print(f"{num} Prime Number nahi hai.")
````

# **Q13:** Nested loop ka use karke niche diya gaya pattern print karein:
```
*
**
***
****
*****
```

````python
# type your code here
# Rows ki sankhya (is case mein 5)
rows = 5

# Outer loop rows ke liye
for i in range(1, rows + 1):
    # Inner loop stars print karne ke liye
    for j in range(1, i + 1):
        print("*", end="") # end="" taaki star ek hi line mein rahe
    
    # Ek row khatam hone ke baad nayi line par jaane ke liye
    print()
````

# **Q14:** Multiplication tables print karein 1 se lekar 5 tak (Nested loops use karke).

````python
# type your code here
for i in range(1, 6):
    print(f"--- Table of {i} ---")
    
    # Inner loop: Yeh har number ko 1 se 10 tak multiply karega
    for j in range(1, 11):
        result = i * j
        print(f"{i} x {j} = {result}")
    
    # Har table ke baad ek khali line (better formatting ke liye)
    print()
````

# **Q15:** Ek list of lists di gayi hai: `matrix = [[1, 2], [3, 4], [5, 6]]`. Nested loop ka use karke saare individual numbers print karein.

````python
# type your code here
matrix = [[1, 2], [3, 4], [5, 6]]

# Outer loop: Har sub-list (row) ko uthayega
for row in matrix:
    # Inner loop: Us sub-list ke har element (number) ko uthayega
    for num in row:
        print(num)
````

# **Q16:** User se ek number input lein aur us ka **Factorial** find karein (e.g., 5! = 5*4*3*2*1).

````python
# type your code here
# User se number input lein
num = int(input("Factorial nikalne ke liye ek number dein: "))

factorial = 1

# Check karein agar number negative hai
if num < 0:
    print("Negative numbers ka factorial nahi hota.")
elif num == 0:
    print("0 ka factorial 1 hota hai.")
else:
    # 1 se lekar num tak multiply karein
    for i in range(1, num + 1):
        factorial *= i
    
    print(f"{num} ka factorial {factorial} hai.")
````
